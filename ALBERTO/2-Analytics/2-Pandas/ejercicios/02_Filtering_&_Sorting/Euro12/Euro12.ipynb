```python
# repaso:

def suma (n,m): 
    n = n + m
    return n
# n=2 y m=3, lo que hago es pasarle una copia de sus valores, por eso el return es 5 (la suma)
# pero si le pregunto por n, me dirá n=2 que es su valor inicial, no el resultado de la función


# a esta función, lo que le paso NO son colecciones, por lo qeu cuando yo la ejecuto,
# lo que hago es pasarle una copia de los valores


def agrega (lista, cosa):
    lista.append(cosa)
    return lista

a = 2 
b = 3

print(suma(a,b)) # 5
print(a) # 2 porque es una copia

a = ["hola"]
b = "adios"
print(agrega(a,b)) # aquí se pasa una lista (que es una colección), es la misma flecha en phyton tutor
print(a) # sale ['hola', 'adios'] porque lo he modificado, la lista se ha agrandado
```

#### Ejercicio 1

Escribe un programa que simule un libro de "sigue tu propia aventura".

Solo se pide control de flujo y uso de print().
El programa tiene dos variables:

- puerta --> es una variable de tipo entero

- amuleto --> es una variable booleana

Si la puerta que eliges es la 1, imprime "has perdido"

Si la puerta que eliges es la 2 y tienes el amuleto (verdadero), imprime "has ganado"

En cualquier otro caso, imprime "sigue jugando"

Haz este programa de dos formas: usando un AND y sin usar un AND

Nota: no son necesarias funciones. Puedes declarar puerta y amuleto al principio de tu programa.

```python
print("=========== JUEGO: SIGUE TU PROPIA AVENTURA ===========")

jugador = input("¿Cuál es tu nombre? ")
amuleto = False

print(f"\nBienvenid@ {jugador}.\nTe encuentras ahora mismo en un frondoso bosque y necesitas refugiarte de la lluvia."
      "\nVes una cueva a pocos metros de ti. ¿Entras dentro? (si/no)")

# programa sin usar AND. 
# Usando AND sería que tuviera el amuleto y la puerta 2, ambas a True

while True:
    avanzar = input().lower().strip()

    if avanzar == "si":
        print(f"\nDejas tu caballo atado en la entrada y enciendes una antorcha para ver el interior.")
        print("Hay un cofre antiguo de madera con termitas. ¿Quieres abrirlo? (si/no)")

        while True:
            abrir = input().lower().strip()

            if abrir == "si":
                amuleto = True
                print(f"\nLo abres y coges un amuleto de jade lleno de polvo."
                      "\nSigues andando y muy adentro encuentras dos puertas numeradas 1 y 2."
                      "\n¿Qué puerta escoges? (1/2)")

                while True:
                    puerta_input = input().strip()

                    try:
                        puerta = int(puerta_input)
                    
                    except ValueError:
                        print("\nSolo puedes contestar con 1 o 2.")
                        continue

                    if puerta == 1:
                        print(f"\nSale un dragón escupiendo fuego y mata a {jugador}.\n\nGAME OVER")
                        break

                    elif puerta == 2:
                        print(f"\nAbres la puerta y ¡consigues el tesoro, {jugador}!\n\nYOU WIN")
                        break

                    else:
                        print("\nSolo puedes contestar con 1 o 2.")
                        continue

                break  # Salimos del bucle de abrir

            elif abrir == "no":
                print(f"\nTe sientas encima a tomarte una hidromiel y mueres borracho.\n\nGAME OVER")
                break

            else:
                print("\nSolo puedes contestar con 'si' o 'no'.")

        break  # Fin del juego después de abrir o no el cofre

    elif avanzar == "no":
        print(f"\nTe quedas fuera bajo la tormenta y mueres de frío.\n\nGAME OVER")
        break

    else:
        print("\nSolo puedes contestar con 'si' o 'no'.")
```

Crea un programa que simule un carrusel infantil.

El carrusel infantil tiene 3 posibles cochecitos dentro de una variable global de tipo lista, fuera de tu función.
- barco 
- bomberos 
- avioneta 

Debes escribir una **FUNCIÓN** llamada carrusel que tenga dos parámetros de entrada: 
- el cochecito que eliges (puede valer el string: barco, bomberos o avioneta) 
- y un segundo parámetro llamado vueltas que es un entero.

Tu función carrusel NO debe retornar nada, pero debe llamar internamente a coloca(cochecito), ya te lo doy programado. 

Si sientas al niño en "avioneta" y pones 3 vueltas, la función debería imprimir:

"vuelta 1: avioneta - barco - bomberos"

"vuelta 2: avioneta - barco - bomberos"

"vuelta 3: avioneta - barco - bomberos"

Se valorará positivamente que haya **UN BUCLE WHILE**.

Si no te acaba de funcionar llamando a coloca(cochecito) tras un tiempo prudencial, escribe tu código sin usarlo.

```python
lista = ["barco", "bomberos", "avioneta"]

def carrusel(coche, vueltas):
    resultado = []

    if coche == "barco":    # estos ifs se harán primero, antes de decirle qué hacer con el FOR
        return lista
    
    elif coche == "bomberos" or coche == "avioneta":
        elemento = lista.pop(0)
        lista.append(elemento)

    if coche == "avioneta":
        elemento = lista.pop(0)
        lista.append(elemento)

    for i in range(vueltas):    # ahora hago los arreglos para que se imprima como quiero
        arreglo = f"vuelta {i + 1}: {' - '.join(lista)}"
        resultado.append(arreglo)
    
    return '\n'.join(resultado)

print(carrusel("avioneta", 3))
```


#### Ejercicio 5
Crea un programa que pida una contraseña por teclado e indique si es correcta
o incorrecta con un print(). 

La contraseña correcta es “python”. 

Si la contraseña era incorrecta la pide de nuevo hasta 3 veces como máximo. 

Si al tercer intento no se acierta, debe ponerse a None y poner en un print() que la cuenta se ha bloqueado. Si se consigue entrar, se muestra que es correcta en un print(). Debe haber como mínimo un **BUCLE WHILE**


```python
password = "python"
intentos = 3

while 0 < intentos <= 3:
    prueba = input("Contraseña: ")
    print(prueba)

    if prueba == password:
        print("Contraseña correcta")
        break
    
    elif prueba != password:
        intentos = intentos - 1
        print(f"Te quedan {intentos} intentos.")
        continue

if intentos == 0:
    intentos = None
    print("Cuenta bloqueada.")
```

```python
class Flor:
    def __init__(self, nombre = str, venenosa = bool, lista_viveros = list):
        self.nombre = nombre
        self.venenosa = venenosa
        self.lista_viveros = lista_viveros

    def info(self):
        if self.venenosa == True:
            self.venenosa = "venenosa"
        else:
            self.venenosa = "no venenosa"
        return f"La flor {self.nombre} es {self.venenosa} y se vende en {len(self.lista_viveros)} viveros."
    
    def cantidad_viveros(self):
        return len(self.lista_viveros)

# Crear una instancia de la clase con flores
rosa = Flor("rosa", False, ["Granada", "Jaén"])
margarita = Flor("margarita", False, ["Huesca", "Zaragoza", "Teruel"])
adelfa = Flor("adelfa", True, ["Madrid"])

# Acceder a los atributos y métodos
print(margarita.info())
print(margarita.cantidad_viveros())
```

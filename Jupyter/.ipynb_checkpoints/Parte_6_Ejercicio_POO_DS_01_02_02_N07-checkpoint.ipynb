## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    tipo = "Electrodomésticos" # aquí se colocan los atributos de clase que son COMUNES a todas las Tiendas
    abierta = True

    def __init__(self, nombre, direccion, empleados, ventas): # aquí se colocan los atributos de instancia que son PROPIOS a cada tienda
        self.nombre = str(nombre)
        self.direccion = str(direccion)
        self.empleados = int(empleados)
        self.ventas = list(ventas)

    # Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
    def ventas_totales(self):
        return sum(self.ventas) 

    # Método que calcula la media de ventas de los ultimos meses, por empleado
    def ventas_empleados(self):
        if self.empleados != 0: # IMP! si voy a sumar un número, tengo que convertirlo a str para concatenarlo
            return "Media ventas por empleado: " + str(round((sum(self.ventas) / self.empleados), 2))
        else:
            return "La tienda no tiene empleados."

    # Método que devuelve en un string el nombre de la tienda, junto con su dirección.
    def ubicacion(self):
        return self.nombre + " se encuentra en: " + self.direccion 
    
    # Método que obtiene las ventas del último mes.
    def ventas_ultimo_mes(self):
        if self.ventas:
            return self.ventas[-1]
        else:
            return 0 # devuelve un 0 si en el último mes no se ha vendido nada de nada
        
    # Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda.     
    # Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen 
    # sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). 
    # El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.
    def proyeccion(self, inversion):
        if inversion < 1000:
            return round((float(sum(self.ventas)) * 1.2), 2) # redondeo a 2 decimales por si acaso me da un número infinito periódico
        elif inversion >= 1000:
            return round((float(sum(self.ventas)) * 1.5), 2) 
```

```python
# 1. Implementar la clase "Tienda"
# 2. Crear tres tiendas con datos inventados
# 3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)

# TIENDA 1
print("\n------------- TIENDA 1 -------------\n")
tienda_1 = Tienda("La tostadora valiente", "Calle Fantasía, 20", 4, [1485,1562,1359])

print(tienda_1.tipo)
print(tienda_1.abierta)
print(tienda_1.nombre)
print(tienda_1.direccion)
print(tienda_1.empleados)
print(tienda_1.ventas)
print(tienda_1.ventas_totales())
print(tienda_1.ventas_empleados())
print(tienda_1.ubicacion())
print(tienda_1.ventas_ultimo_mes())
print(tienda_1.proyeccion(700))

# TIENDA 2
print("\n------------- TIENDA 2 -------------\n")
tienda_2 = Tienda("Risas y voltios", "Avenida Me falta un tornillo, 71", 9, [7519,7210,6148])

print(tienda_2.tipo)
print(tienda_2.abierta)
print(tienda_2.nombre)
print(tienda_2.direccion)
print(tienda_2.empleados)
print(tienda_2.ventas)
print(tienda_2.ventas_totales())
print(tienda_2.ventas_empleados())
print(tienda_2.ubicacion())
print(tienda_2.ventas_ultimo_mes())
print(tienda_2.proyeccion(1000))

# TIENDA 3
print("\n------------- TIENDA 3 -------------\n")
tienda_3 = Tienda("Horno caliente", "Callejón Sal si puedes, 6", 12, [8411,9074,9478])

print(tienda_3.tipo)
print(tienda_3.abierta)
print(tienda_3.nombre)
print(tienda_3.direccion)
print(tienda_3.empleados)
print(tienda_3.ventas)
print(tienda_3.ventas_totales())
print(tienda_3.ventas_empleados())
print(tienda_3.ubicacion())
print(tienda_3.ventas_ultimo_mes())
print(tienda_3.proyeccion(1200))

# 4. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle for
print("\n------------------------------------\n")
lista_tiendas = [tienda_1, tienda_2, tienda_3]
for tienda in lista_tiendas:
    print("Las ventas de", tienda.nombre, "durante el último mes han sido:", tienda.ventas_ultimo_mes())

# 5. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
print("\n------------------------------------\n")
palabra = "Avenida"
for tienda in lista_tiendas:
    if palabra in tienda.direccion.split(" "):
        print("La palabra", palabra, "está en la dirección de la tienda:", tienda.nombre)

# 6. Seguro que ya has hecho este apartado... Documenta la clase :)
print("\n------------------------------------\n")
print("La clase 'Tienda' es de tipo:", type(Tienda))
for tienda in lista_tiendas:
    print("=======", tienda.nombre, "=======")
    print("Tipo:", type(tienda.tipo))
    print("Abierta:", type(tienda.abierta))
    print("Nombre:", type(tienda.nombre))
    print("Dirección:", type(tienda.direccion))
    print("Empleados:", type(tienda.empleados))
    print("Ventas:", type(tienda.ventas))
    print("Ventas totales:", type(tienda.ventas_totales()))
    print("Ventas empleados:", type(tienda.ventas_empleados()))
    print("Ubicación:", type(tienda.ubicacion()))
    print("Ventas último mes:", type(tienda.ventas_ultimo_mes()))
    print(tienda.proyeccion(850)) # pongo 850€ para que salga por pantalla sin errores, aunque cada una de ellas tiene su propia inversión
    print("\n")
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0 # el perro está parado

    def __init__(self, raza, color = "marrón", propietario = False):
        self.raza = str(raza)
        self.color = str(color)
        self.propietario = propietario

    def andar(self):
        try: # pongo try/except por si mete un dato no válido en el input, que indique que es erróneo
            aumento_velocidad = float(input("Aumento de velocidad:"))
            self.velocidad = aumento_velocidad + self.velocidad
            return self.velocidad
        except ValueError:
            print("Introduce un número válido.")
            return self.velocidad

    def parar(self):
        print("El perro se detiene.")
        self.velocidad = 0
        return self.velocidad

    def ladrar(self):
        nombre = input("¿Cuál es el nombre del perro?:")
        return "GUAU! " + nombre
```

```python
perro_1 = Perro("husky")

print("------ MI PERRITO FALDERO -----")
print("Patas:", perro_1.patas)
print("Orejas:", perro_1.orejas)
print("Ojos:", perro_1.ojos)
print("Velocidad:", perro_1.velocidad)
print("Raza:", perro_1.raza)
print("Color:", perro_1.color)
print("Propietario:", perro_1.propietario)

print("\n")
print("Andar:", perro_1.andar())
print("La velocidad actual del perro es:", perro_1.velocidad)
print("Si ahora se detiene:", perro_1.parar())
print("Su velocidad pasaría a ser:", perro_1.velocidad)

print("\n")
print("Ladrar:", perro_1.ladrar())
print("La clase 'Perro' es de tipo:", type(Perro))
```
